# Read

```python
ratio_comb_df = pd.read_excel(path + '/3-result/ratio_combinations.xlsx', index_col='Unnamed: 0')
ratio_comb_df.fillna(0, inplace=True)
ratio_comb_df.sort_values(by='oran', ascending=False, inplace=True, ignore_index=True)
ratio_comb_df = ratio_comb_df.astype({'saat1':int,'saat2':int,'süre1':int,'süre2':int})

del ratio_comb_df['freq']
ratio_comb_df = ratio_comb_df.iloc[4:]

ratio_comb_df.index = range(1,ratio_comb_df.shape[0]+1)

## 

roll_comb_df = pd.read_excel(path + '/3-result/roll_combinations.xlsx', index_col='Unnamed: 0')
roll_comb_df.fillna(0, inplace=True)
roll_comb_df.sort_values(by='katılım', ascending=False, inplace=True, ignore_index=True)
roll_comb_df = roll_comb_df.astype({'saat1':int,'saat2':int,'süre1':int,'süre2':int})

roll_comb_df = roll_comb_df.iloc[4:]

roll_comb_df.index = range(1,roll_comb_df.shape[0]+1)
```

```python
def styler(df, sort):
    def color(val):
        return 'background-color: %s' % ('#f29797' if val == 'Farklı' else '#adf5ad')


    s = df.style
    s.applymap(color, subset=['gün', 'saat', 'süre'])
    s.background_gradient(subset=[sort])
    s.format({"saat": lambda x: f"{x}:00" if x != 'Farklı' else x,
            "süre": lambda x: f"{x} saat" if x != 'Farklı' else x})

    return s
```

# Format Roll

```python
format_roll_comb_df = roll_comb_df[roll_comb_df.columns[:3]]

with warnings.catch_warnings(record=True):
    warnings.simplefilter("always")

    empty_ = np.empty(roll_comb_df.shape[0], dtype="<U9")
    empty_[roll_comb_df['gün1'] == roll_comb_df['gün2']] = roll_comb_df['gün1'][roll_comb_df['gün1'] == roll_comb_df['gün2']]
    empty_[roll_comb_df['gün1'] != roll_comb_df['gün2']] = 'Farklı'
    format_roll_comb_df['gün'] = empty_

    empty_ = np.empty(roll_comb_df.shape[0], dtype="<U9")
    empty_[roll_comb_df['saat1'] == roll_comb_df['saat2']] = roll_comb_df['saat1'][roll_comb_df['saat2'] == roll_comb_df['saat1']]
    empty_[roll_comb_df['saat1'] != roll_comb_df['saat2']] = 'Farklı'
    format_roll_comb_df['saat'] = empty_

    empty_ = np.empty(roll_comb_df.shape[0], dtype="<U9")
    empty_[roll_comb_df['süre1'] == roll_comb_df['süre2']] = roll_comb_df['süre1'][roll_comb_df['süre1'] == roll_comb_df['süre2']]
    empty_[roll_comb_df['süre1'] != roll_comb_df['süre2']] = 'Farklı'
    format_roll_comb_df['süre'] = empty_
```

```python
styler(format_roll_comb_df, 'katılım')
```
